## Cleaning

```python
backup=cars.copy()
backup.shape
```

```python
# Cleaning year------------------------------------------------------------------

cars=cars[cars['year'].str.isnumeric()]
cars.shape
```

```python
cars['year']=cars['year'].astype(int)
cars.info()
```

```python
# clerning price------------------------------------------------------------------

cars=cars[cars['Price'] != 'Ask For Price']
cars.head()
```

```python
cars['Price'] = cars['Price'].str.replace(',', '').astype(int)
cars.head(2)
```

```python
cars.info()
```

```python
# cleaning kms drivcen------------------------------------------------------------------

cars=cars[cars['kms_driven'].str.contains('km') & cars['kms_driven'].str.contains('km')]
cars.head()
cars.shape
```

```python
cars['kms_driven']=cars['kms_driven'].str.replace(',','').replace('km','').str.replace(r'kms|km','',regex=True).str.strip().astype(int)
```

```python
cars.info()
```

```python
cars=cars[~cars['fuel_type'].isna()]
```

```python
cars.head()
```

```python
cars['name']=cars['name'].str.split().str[:3].str.join(' ')
```

```python
cars.reset_index(drop=True,inplace=True)
cars.head()
```

```python
cars.describe()
```

```python
cars=cars[cars['Price']<=6e6].reset_index(drop=True)
```

```python
cars.to_csv('cleaned_car.csv')
```

# Model

```python
cars.head()
```

```python
x=cars.drop('Price',axis=1)
y=cars['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

ohe=OneHotEncoder()
ohe.fit(x[['name', 'company', 'fuel_type']])
columns_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name', 'company', 'fuel_type']),remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

model=LinearRegression()
# pipe=make_pipeline(columns_trans,model)
```

```python
score=[]
for i in range(1000):
    x_train2,x_test2,y_train2,y_test2 = train_test_split(x,y,test_size=0.2,random_state=i)
    pipe=make_pipeline(columns_trans,model)
    pipe.fit(x_train2,y_train2)
    y_pred2=pipe.predict(x_test2)
    score.append(r2_score(y_test2,y_pred2))
    
```

```python
from matplotlib import pyplot as plt
states = np.arange(0, 1000)
plt.figure(figsize=(10, 6))
plt.plot(states, score, color='tab:blue')
plt.title('Price vs Kms Driven (Line Graph)')
plt.xlabel('Sorted Kms Driven')
plt.ylabel('Sorted Price')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
max_state=np.argmax(score)
x_train2,x_test2,y_train2,y_test2 = train_test_split(x,y,test_size=0.2,random_state=max_state)
pipe=make_pipeline(columns_trans,model)
pipe.fit(x_train2,y_train2)
y_pred2=pipe.predict(x_test2)
r_final=r2_score(y_test2,y_pred2)
r_final
```

```python
import pickle
pickle.dump(pipe,open('car_price_model.pkl','wb'))
```

```python
model = pickle.load(open('car_price_model.pkl', 'rb'))
test_car = pd.DataFrame([['Maruti Suzuki Swift', 'Maruti', 2019, 15000, 'Petrol']], 
                        columns=['name', 'company', 'year', 'kms_driven', 'fuel_type'])
predicted_price = model.predict(test_car)
print(f"The model predicts this car's price is: ₹{predicted_price[0]:,.2f}")
```
